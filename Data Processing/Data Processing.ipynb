```python
!pip3 install openpyxl 
!pip3 install matplotlib
import os
import openpyxl as xl
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import plotly.express as px

Collated =  "/home/dmacs/Documents/Pranav_Phanindra/Data_Scraping_Final/Data_Part_2/Collated"


DataFiles = os.listdir(Collated)
DataFiles.sort()
DataFiles
```

```python
from openpyxl import load_workbook

def get_sheet_names(filepath) :
    wb = load_workbook(filepath , read_only = True , keep_links = False)
    return wb.sheetnames
```

```python
sheets = get_sheet_names(Collated + '/{}'.format(DataFiles[0]))
sheets
```

# Mustard

```python
main_dataframe = pd.DataFrame()

for sheet in sheets :
    df = pd.read_excel(open(Collated+'/{}'.format(DataFiles[0]), 'rb'), sheet_name=sheet)
    if len(df.columns) != 1 and 'Modal Price (Rs./Quintal)' in df.columns :
        df = df.iloc[:,2:]
        df['State'] = sheet
        
        main_dataframe = pd.concat([main_dataframe , df] , axis = 0)
        print('{}---{}---{}'.format('yes',DataFiles[0],sheet))

main_dataframe
```

```python
main_dataframe['Yr-mon'] = main_dataframe['Price Date'].dt.strftime('%Y%m')
mf = main_dataframe.groupby(['Yr-mon'],as_index = False).mean()
mf
```

```python
ms = main_dataframe.groupby(['State','Yr-mon'] ,  as_index = False ).mean()
ms
```

```python
fig = px.line(ms, x='Yr-mon', y='Modal Price (Rs./Quintal)', color= 'State')
fig.show()
fig.write_html(Collated + "/Graphs/State_Mustard.html")
```

```python
main_dataframe['Year'] = pd.DatetimeIndex(main_dataframe['Price Date']).year
main_dataframe['Month'] = pd.DatetimeIndex(main_dataframe['Price Date']).month
main_dataframe
```

```python
main_dataframe.groupby(['Year','Month']).mean()
```
