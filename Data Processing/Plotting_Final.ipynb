```python
!pip3 install openpyxl 
!pip3 install matplotlib
import os
import openpyxl as xl
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import plotly.express as px

Collated =  "/home/dmacs/Documents/Pranav_Phanindra/Data_Scraping_Final/Data_Part_2/Collated"


DataFiles = os.listdir(Collated)
DataFiles.sort()
print(DataFiles)

def get_sheet_names(filepath) :
    wb = xl.load_workbook(filepath , read_only = True , keep_links = False)
    return wb.sheetnames
```

```python
sheets = get_sheet_names(Collated + '/{}'.format(DataFiles[0]))
sheets
```

```python
for file in DataFiles :
    main_dataframe = pd.DataFrame()
    ms = pd.DataFrame()
    mf = pd.DataFrame()
    
    for sheet in sheets :
        df = pd.read_excel(open(Collated+'/{}'.format(file), 'rb'), sheet_name=sheet)
        if len(df.columns) != 1 and 'Modal Price (Rs./Quintal)' in df.columns :
            df = df.iloc[:,2:]
            df['State'] = sheet
            
            main_dataframe = pd.concat([main_dataframe , df] , axis = 0)
            print('{}---{}---{}'.format('yes',file,sheet))
            
            main_dataframe['Yr-mon'] = main_dataframe['Price Date'].dt.strftime('%Y%m')
            mf = main_dataframe.groupby(['Yr-mon'],as_index = False).mean()
            ms = main_dataframe.groupby(['State','Yr-mon'] ,  as_index = False ).mean()
    
    #Statewise Graphs
    fig = px.line(ms, 
                  x='Yr-mon', 
                  y='Modal Price (Rs./Quintal)', 
                  color= 'State' , 
                  title = "Statewise {}".format(file.split('_')[1].split('.')[0])
                 )
    fig.write_html(Collated + "/State_Graphs/State_{}.html".format(file.split('_')[1].split('.')[0]))
    print("State_{}.html".format(file.split('_')[1].split('.')[0]),
          '\t' ,
          'done')
    
    #Commodity wise full graphs 
    fig = px.line(mf, 
                  x='Yr-mon', 
                  y='Modal Price (Rs./Quintal)',
                  title =  "{}".format(file.split('_')[1].split('.')[0])
                 )
    fig.write_html(Collated + "/Commodity_Graphs/{}.html".format(file.split('_')[1].split('.')[0]))
    print("{}.html".format(file.split('_')[1].split('.')[0]) ,
          '\t' , 
          'done')
        
```
