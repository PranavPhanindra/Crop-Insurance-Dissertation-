```python
import pandas as pd
import os 
```

```python
src_path = r'/home/dmacs/Documents/Pranav_Phanindra/Data_Scraping_Final/Data_Part_2/Collated/CSVs/Excel/'
dest_path = r'/home/dmacs/Documents/Pranav_Phanindra/Data_Scraping_Final/Data_Part_2/Collated/Prices_Aggregated/'
```

```python
files = os.listdir(src_path)
files
```

```python
filename = files[-1]
filename
```

```python
df = pd.read_excel(src_path+filename)
df
```

```python
df['Price Date'] =  pd.to_datetime(df['Price Date'])
df['Yr-mon'] = df['Price Date'].dt.strftime('%Y%m')
df['Yr-mon'] = pd.to_datetime(df['Yr-mon'], format = '%Y%m')

df['Year'] = df.set_index('Yr-mon').index.year
df['Month'] = df.set_index('Yr-mon').index.month
df.set_index('Price Date')
```

```python
df['Year'].unique()
```

```python
df_aggregated = df.groupby(['Yr-mon'],as_index = False).mean()
df_aggregated
```

```python
df_aggregated['Year'] = df_aggregated.set_index('Yr-mon').index.year
df_aggregated['Month'] = df_aggregated.set_index('Yr-mon').index.month
df_aggregated
```

```python
df_aggregated.to_excel(dest_path+'{}_Aggregated.xlsx'.format(filename.split('.')[0]))
```
