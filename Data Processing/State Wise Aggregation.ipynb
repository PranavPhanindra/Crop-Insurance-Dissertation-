```python
import pandas as pd
import os
```

```python
src_path = r'/home/dmacs/Documents/Final Dissertation/Data/Prices/Individual Collated/Excel/'
dest_path = r'/home/dmacs/Documents/Final Dissertation/Data/Prices/Individual Collated/StateWise_Aggregated/'
```

```python
files = os.listdir(src_path)
files
```

```python
filename = files[6]
filename
```

```python
df = pd.read_excel(src_path+filename)
df
```

```python
df['Price Date'] =  pd.to_datetime(df['Price Date'])
df['Yr-mon'] = df['Price Date'].dt.strftime('%Y%m')
df['Yr-mon'] = pd.to_datetime(df['Yr-mon'], format = '%Y%m')
df['Year'] = df.set_index('Yr-mon').index.year
df['Month'] = df.set_index('Yr-mon').index.month
df
```

```python
statewise_aggregated = df.groupby(['Yr-mon','State'],as_index = False).mean()

statewise_aggregated['Year'] = statewise_aggregated.set_index('Yr-mon').index.year
statewise_aggregated['Month'] = statewise_aggregated.set_index('Yr-mon').index.month
statewise_aggregated
```

```python
dest_filename = filename.split('.')[0]
dest_filename
```

```python
statewise_aggregated.to_excel(dest_path+'{}_statewise_aggregated.xlsx'.format(dest_filename))
```
