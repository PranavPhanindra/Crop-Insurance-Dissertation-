```python
import pandas as pd
```

```python
prices = pd.read_excel(r'/home/dmacs/Documents/Final Dissertation/Data/Statewise Prices.xlsx')
volumes = pd.read_excel(r'/home/dmacs/Documents/Final Dissertation/Data/Statewise Volumes.xlsx')
```

```python
merged_df = pd.merge(prices, volumes,on=['UID'])
merged_df
```
