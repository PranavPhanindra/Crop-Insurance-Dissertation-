```python
!pip3 install openpyxl 
from pathlib import Path

import os
import openpyxl as xl
import pandas as pd

Collated =  Path("/home/dmacs/Documents/Pranav_Phanindra/Data_Scraping_Final/Data_Part_2/Collated")

```

```python
DataFiles = os.listdir(Collated)
DataFiles.sort()
print(DataFiles)
```

```python
def get_sheet_names(filepath) :
    wb = xl.load_workbook(filepath , read_only = True , keep_links = False)
    return wb.sheetnames
```

```python
sheets = get_sheet_names(Collated + '/{}'.format(DataFiles[0]))
sheets
```

```python
for file in DataFiles :
    if 'xlsx' in file.split('.') :
        main_dataframe = pd.DataFrame()
        sheets = get_sheet_names(Collated  /  '{}'.format(file))
        for sheet in sheets :
            df = pd.read_excel(open(Collated  / '{}'.format(file), 'rb'), sheet_name=sheet)
            if len(df.columns) != 1 and 'Modal Price (Rs./Quintal)' in df.columns :
                print(file,'--',sheet,'-- ','true')
                df = df.iloc[:,2:]
                df['State'] = sheet
                print()
                main_dataframe = pd.concat([main_dataframe , df] , axis = 0)
            
            
        main_dataframe.to_csv(Collated  / '{}.csv'.format(file.split('.')[0]) )
```
