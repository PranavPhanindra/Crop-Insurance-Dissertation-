```python



for commodity_index , commodity_name  in commodity_indices.items() :
    
    
    for state_index , state_name  in states_indices.items() :
        state_name = '+'.join(state_name.split(' '))
        
        
        #Setting preferences to download directory
        profile = webdriver.FirefoxProfile()
        profile.set_preference('browser.download.folderList',2)
        profile.set_preference('browser.download.manager.showWhenStarting',False)
        profile.set_preference('browser.download.dir',
                                   '/home/dmacs/Documents/Pranav_Phanindra/Data_Scraping_Final/Data_Part_2/Data_excel_2/')
        profile.set_preference('browser.helperApps.neverAsk.saveTODisk','application/x-gzip')
        
        
        link = "https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity={}&Tx_State={}&Tx_District=0&Tx_Market=0&DateFrom=01-Apr-2008&DateTo=31-Mar-2022&Fr_Date=01-Apr-2008&To_Date=31-Mar-2022&Tx_Trend=0&Tx_CommodityHead={}&Tx_StateHead={}&Tx_DistrictHead=--Select--&Tx_MarketHead=--Select--".format(commodity_index , state_index ,commodity_name,state_name)
        
        print(link,'\n')
        
        link_driver = webdriver.Firefox(capabilities=capabilities,firefox_profile=profile)
        link_driver.get(link)
        
        time.sleep(5)
        
        try:
        #Wait until the required element is seen 
            element = WebDriverWait(link_driver, 10) 
            element.until(EC.presence_of_element_located((By.ID, "cphBody_ButtonExcel")))
        
            state = link_driver.find_element(By.XPATH , "//*[@id='ddlState']").get_attribute(name='value')
            commodity = link_driver.find_element(By.ID , 'ddlCommodity').get_attribute(name='value')


            #Trying to dowload data in excel format
            to_Excel = link_driver.find_element(By.XPATH , "//*[@id='cphBody_ButtonExcel']")
            to_Excel.click()
            #time.sleep(10)

            path = r"/home/dmacs/Documents/Pranav_Phanindra/Data_Scraping_Final/Data_Part_2/Data_excel_2"
            files = os.listdir(path)
            filename = "Agmarknet_Price_Report.xls"
            present = True

            while  present :
                if filename in files :
                    print("Downloaded...")
                    time.sleep(5)
                    
                    present = False
                    break
                else :
                    print("Downloading...")
                    time.sleep(5)

                files = os.listdir(path)


                    
            
            old_name =  old_path + 'Agmarknet_Price_Report.xls'
            new_name_1 = new_path_1 + '{}_{}.xls'.format(commodity,state)
            new_name_2 = new_path_2 + '{}_{}.html'.format(commodity,state)

            if 'Agmarknet_Price_Report.xls' in os.listdir(old_path) :
                shutil.copy(old_name, new_name_1)
                os.rename(old_name, new_name_2)
    
    
        finally:
            link_driver.quit()
            print('Done')
```

```python
for file in files :
    #Source excel sheet and source workseheet in excel 
    source_xl = openpyxl.load_workbook(excel_path_to+'/{}'.format(file))
    source_sheet = source_xl.worksheets[0]
    
    #Store sheetname to load the data
    sheetName = states_indices[file.split('_')[1].split('.')[0]]
    commodityIndex = file.split('_')[0]
    
    #Get the file name to be loaded as workbook
    xlFileName = '{}_{}.xlsx'.format(commodityIndex,commodity_indices[int(commodityIndex)])
    
    #Loading the destination workbook and worksheet
    workbook = openpyxl.load_workbook(collated+'/{}'.format(xlFileName))
    worksheet = workbook.get_sheet_by_name(sheetName)
    
    mr = source_sheet.max_row
    mc = source_sheet.max_column
    for i in range(1,mr+1) :
        for j in range(2,mc+1) :
            
            val = source_sheet.cell(row = i, column = j)
            worksheet.cell(row = i, column = j).value = val.value
            
    workbook.save(collated+'/{}'.format(xlFileName))
    print('Done','\t',file,'\t',xlFileName,'\t')
    source_xl.close()
    workbook.close()
            
```
